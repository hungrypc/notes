# Monte Carlo simulations as a decision making tool

When we run a Monte Carlo simulation, we are interested in observing the different possible realizations of a future event. We can use past data to create a simulation, a new set of fictional but sensible data. These realizations are generated by observing the distribution of the historical data and calculating its mean and variance. The new data set is larger than the initial set and it allows us to see what would have happened.

This is valuable as it allows us to consider a good proxy of the probability of different outcomes and can help us make an informed decision. 


## Monte Carlo applied in a Corporate Finance context

Current Revenues = Last Year Revenues * (1 + y-o-y growth rate)

Last year's revenues are available. The random variable in this case is revenue growth rate. A computer software would allow us to run 1000 simulations of the development of revenues to obtain a maximum minimum and average value of the expected revenues figure. 

The two parameters, revenue growth and standard deviation (revenue volatility) could be obtained by looking at historical figures or arbitrarily chosen based on the user's intuition. 

The reasoning for cost of goods sold (cogs) and operating expenses (opex) is almost the same with one main difference. We need to represent cogs and opex as a percentage of revenues, then model the development of the percentage of revenues. This would allow us to obtain cogs and opex figures, which are in line with the firm's expected revenues as cogs and opex depend on the revenues the firm makes. If revenues are low, there'll be no need to spend a lot for cogs as fewer products will be produced. 

When you deduct cogs from the revenue of a company, you will obtain its gross profit.

## Predicting Gross Profit

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rev_m = 170        # revenue mean
rev_stdev = 20     # revenue standard dev.
iterations = 1000
```

```python
# simulations of future revenues
rev = np.random.normal(rev_m, rev_stdev, iterations)
rev
```

```python
plt.figure(figsize=(15,6))
plt.plot(rev)
plt.show()
```

We can see that the simulated values are centered around the mean of 170, and that almost all data points fall between 150~190 (170 +/- 20). 

Assume that you're experienced in this business and are able to tell that typically COGS amount to approx. 60% of the revenues. 60% of the revenues are spent to produce the goods it sells. The estimation of 60% will have a standard dev. of 10%. COGS is money spent, therefore we should put a minus sign before the expression reflects the multiplication of revenues by 60%. 

```python
cogs = - (rev * np.random.normal(0.6,0.1))

plt.figure(figsize=(15, 6))
plt.plot(cogs)
plt.show()
```

```python
cogs.mean()
```

```python
cogs.std()
```

The difference between reveues and cogs gives us gross profit. Computing the latter is the objective. Since we've generated 1000 potential revenue and cogs values, calculating the simulation of gross profit requires us to combine the revenue and cogs. 

```python
gross_profit = rev + cogs
gross_profit
```

```python
plt.figure(figsize=(15,6))
plt.plot(gross_profit)
plt.show()
```

It is normally distributed with mean value equal to the difference between the mean of the revenues and the mean of cogs. With the help of max and min functions, it's straightforward to obtain the biggest and smallest potetial value of gross profit. 

```python
max(gross_profit)
```

```python
min(gross_profit)
```

```python
gross_profit.mean()
```

```python
gross_profit.std()
```

Let's plot these results on a histogram. It helps us identify the distribution of our output. 

```python
plt.figure(figsize=(10,6))
# bins are the chunks in which the data in the plot will be divided
# two ways to use bins:
# 1. create a list whose elements separate the bins along the x axis
plt.hist(gross_profit, bins=[40,50,60,70,80,90,100,110,120])
plt.show()
```

```python
# 2. assign the number of bins directly.
plt.figure(figsize=(10,6))
# if i ask for 20 bins, this will split the data into 20 equal intervals
plt.hist(gross_profit, bins=20)
plt.show()
```

## Forecasting Stock Prices with a Monte Carlo Simulation

$$ Price Today = Price Yesterday * e^r $$

r = the log return of the share

This allows us to depict today's stock price as a function of yesterday's stock price and the daily return we'll have. 

Brownian motion is a concept that would allow us to model such randomness. The formula we can use is made of two components:

1. Drift
    - The direction rates of return have been headed in the past
    - Is the best approximation about the future we have
    - = average daily returns - 0.5 * variance
    - Drift is the expected daily return of the stock

2. Volatility
    - The random variable
    - Given by stocks historical volatility * Z(rand(0;1))

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as  plt
from scipy.stats import norm
%matplotlib inline
```

```python
ticker = 'PG'
data = pd.DataFrame()
data[ticker] = wb.DataReader(ticker, data_source ='yahoo', start='2010-1-1')['Adj Close']
```

```python
# we want to forecast PG's future stock price, so first we'll estimate
# its historical log returns.
# here, we use pct.change(), which obtains simple returns from a provided dataset
log_returns = np.log(1 + data.pct_change())
```

```python
log_returns.tail()
```

```python
data.plot(figsize=(10,6))
```

```python
log_returns.plot(figsize=(10,6))
```

```python
u = log_returns.mean()
u
```

```python
var = log_returns.var()
var
```

$$drift = u - \frac{1}{2} * var$$

```python
drift = u - (0.5 * var)
drift
```

```python
stdev = log_returns.std()
stdev
```

```python
np.array(drift)
```

```python
drift.values  # does the same as np.array()
```

```python
stdev.values
```

```python
norm.ppf(0.95)
# this allows us to obtain z
# z corresponds to the distance between the mean and the events, 
# expressed as the number of standard deviations
```

```python
x = np.random.rand(10, 2)
x
```

```python
norm.ppf(x)
```

```python
z = norm.ppf(np.random.rand(10, 2))
z
```

```python
t_intervals = 1000
iterations = 10
```

$$daily returns = e^r$$
$$r = drift + stdev * z$$

```python
daily_returns = np.exp(drift.values + stdev.values * norm.ppf(np.random.rand(t_intervals, iterations)))
```

```python
daily_returns
```

So, the formula we used would allow us to obtain a 1000 by 10 array with daily return values. 10 sets of 1000 random future stock prices. All we have to do now is create a price list.

$$S_t = S_0 * dailyreturn_t$$

$$S_{t+1} = S_t * dailyreturn_{t+1}$$

$$...$$

$$S_{t+999} = S_{t_998} * dailyreturn_{t+999}$$

Each price must equal the profuct of the price observed the previous day and the simulated daily return. Therefore, once we obtain the price and date t, we can estimate the expected stock price we will have in day t + 1. Then, this process will be repeated 1000 times and we will obtain a prediction of a company's stock price 1000 days from now. 

```python
# we start with today's stock price
S0 = data.iloc[-1]
S0
```

```python
price_list = np.zeros_like(daily_returns)
```

```python
price_list
```

Why did we create these zeros? We can replace the zeros with the expected stock price by using a loop.

```python
price_list[0] = S0
price_list
```

```python
for t in range(1, t_intervals):
    price_list[t] = price_list[t - 1] * daily_returns[t]
```

```python
price_list
```

```python
plt.figure(figsize=(10,6))
plt.plot(price_list)
```
