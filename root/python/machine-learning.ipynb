# Machine Learning & Data Science

3 Categories in ML:
1. Supervised Learning
    - Data already has categories/labels
    - Classification: based on inputs, classify something
    - Regression: based on inputs, predict something
2. Unsupervised Learning
    - Data without categories/labels
    - Clustering: creating groups to classify
    - Association Rule Learning: associate different things to predict something
3. Reinforcement
    - All about teaching through trial and error
    - Skill acquisition
    - Real time learning

## Steps for ML:
1. Import the data
2. Clean the data
3. Split the data. Training set/Test set
4. Create a Model
5. Check the output
6. Improve
